```python
# load the data - new data that was provided
supervised_data = pd.read_csv(r"pul_seq_low_high_substr_year_corrected.csv")
```

```python
supervised_strings = supervised_data["sig_gene_seq"]
```

```python
from py_stringmatching.similarity_measure.monge_elkan import MongeElkan
```

```python
def pair_sim(a, b): 
    me = MongeElkan()
    split_a = a.split(",")
    split_b = b.split(",")
    score = me.get_raw_score(split_a, split_b)
    return score
```

```python
from joblib import Parallel, delayed
```

```python
positions = [[i,j] for i in range(0,len(supervised_strings)) for j in range(0,len(supervised_strings))]
```

```python
similarities = Parallel(n_jobs=6, verbose = 5)(delayed(pair_sim)(i, j) for i in supervised_strings for j in supervised_strings)
```

```python
import numpy as np
```

```python
similarities_matrix = np.array(similarities).reshape(len(supervised_strings),len(supervised_strings))
```

```python
# sim_matrix_train = pd.DataFrame(similarities_matrix, columns = y_train, index = y_train)
```

```python
# similarities_matrix
```

```python
mappings = dict(zip(range(0, len(supervised_strings)), supervised_data["high_level_substr"]))
```

```python
similarities_matrix
```

```python
ascending_order = similarities_matrix.argsort()
```

```python
descending_order = ascending_order[:,::-1]
```

```python
descending_order = pd.DataFrame(descending_order)
```

```python
from tqdm.notebook import tqdm
```

```python
for idx in tqdm(mappings.keys()):
    descending_order = descending_order.replace(idx, mappings[idx])
```

```python
descending_order.index = supervised_data["high_level_substr"]
```

```python
descending_order
```

```python
top_5_preds = descending_order.iloc[:,1:6]
```

```python
correct_preds = np.repeat(descending_order.index.tolist(), 5).reshape(654,5)
```

```python
correct_preds == top_5_preds
```

```python
accuracies = (correct_preds == top_5_preds).mean(1)
```

```python
accuracies
```

```python
accuracies = accuracies.reset_index()
```

```python
accuracies.columns = ["high_level_substr", "avg_top_5"]
```

```python
avg_top_5_acc = pd.DataFrame(accuracies.groupby("high_level_substr")["avg_top_5"].mean())
```

```python
avg_top_5_acc
```

```python
# sorted by frequency
avg_top_5_acc.reindex(supervised_data["high_level_substr"].value_counts().keys().tolist())
```

```python
top_5_hist_data = accuracies[accuracies["high_level_substr"].isin(supervised_data["high_level_substr"].value_counts().keys().tolist()[:5])]
```
